Let's begin looking at SVD

```python
plt.figure(figsize=(8, 8))
G = nx.generators.community.ring_of_cliques(15, 10)
nx.draw_spring(G, node_color='magenta', alpha=0.5)
```

```python
A = nx.adjacency_matrix(G).toarray()
plt.imshow(A)
```

```python
U, s, Vh = np.linalg.svd(A)
S = np.diag(s)
PrettyMatrix(s)
```

```python
R = U.dot(S).dot(Vh)
plt.imshow(R)
```

```python
model = TruncatedSVD(n_components=1)
plt.title('Top-k singular vectors. {}'.format(1))
T = model.fit_transform(A)
R = model.inverse_transform(T)
plt.imshow(R)
```

```python
model = TruncatedSVD(n_components=14)
plt.title('Top-k singular vectors. {}'.format(14))
T = model.fit_transform(A)
R = model.inverse_transform(T)
plt.imshow(R)
```

```python
model = TruncatedSVD(n_components=15)
plt.title('Top-k singular vectors. {}'.format(15))
T = model.fit_transform(A)
R = model.inverse_transform(T)
plt.imshow(R)
```

```python
from sklearn.decomposition import TruncatedSVD
def plot(k):
    plt.cla()
    model = TruncatedSVD(n_components=k+1)
    plt.title('Top-k singular vectors. {}'.format(k+1))
    T = model.fit_transform(A)
    R = model.inverse_transform(T)
    plt.imshow(R)
    
fig = plt.figure()
ani = animation.FuncAnimation(fig, plot, len(A)-1)
HTML(ani.to_html5_video())
```

# Now let's look at the bipartite hashtags graph

```python
df = pd.read_csv('./fixed_comments.csv', sep='\t')
del df['Unnamed: 0']
df.loc[:10]
```

```python
df['created_time'] = pd.to_datetime(df['created_time'] * 1e9) # convert to nanoseconds
df = df.set_index('created_time')
df.index = df.index.floor('1H') # round timestamps to the hour
df['time'] = df.index
```

Hacky python code follows. Simply count user to tag mentions. Libraries like spacy, nltk and gensim do this for you.

```python
from collections import defaultdict

users = {}
tags = {}
time = {}
mentions = defaultdict(int)
temporal_mentions = defaultdict(int)
pop_tags = defaultdict(int)
pop_users = defaultdict(int)

for user, tags_str, ts in df[['owner_username', 'tags', 'time']].values:
    for tag in tags_str[1:-1].replace("'", "").strip().split(','):
        if tag.strip():
            for cleaned in tag.split('#'):
                cleaned = cleaned.strip()
                if user not in users:
                    users[user] = len(users)
                if cleaned not in tags:
                    tags[cleaned] = len(tags)
                if ts not in time:
                    time[ts] = len(time)
                mentions[user, cleaned] += 1
                temporal_mentions[user, cleaned, ts] += 1
                pop_tags[cleaned] += 1
                pop_users[user] += 1
                
inverse_users = dict((v, k) for k, v in users.items())
inverse_tags = dict((v, k) for k, v in tags.items())
inverse_time = dict((v, k) for k, v in time.items())
```

Now, let's create a sparse matrix with values in log (remove skew)

```python
X = sp.lil_matrix((len(users), len(tags)))
for user, tag in mentions:
    X[users[user], tags[tag]] = np.log10(mentions[user, tag] + 1)
X = X.tocsr()
```

```python
def print_top_words(model, feature_names):
    for topic_idx, topic in enumerate(model.components_):
        message = "Topic #%d: " % topic_idx
        message += " ".join([feature_names[i]
                             for i in topic.argsort()[:-10 - 1:-1]])
        print(message)
    print()
```

```python
model = TruncatedSVD(n_components=20)
model.fit_transform(X)
print_top_words(model, inverse_tags)
```

```python
model = NMF(n_components=20)
model.fit_transform(X)
print_top_words(model, inverse_tags)
```

```python
filtered_users = {}
filtered_tags = {}
filtered_mentions = defaultdict(int)
for user, tag, ts in temporal_mentions:
    if pop_users[user] >= 5 and pop_tags[tag] >= 5:
        if user not in filtered_users:
            filtered_users[user] = len(filtered_users)
        if tag not in filtered_tags:
            filtered_tags[tag] = len(filtered_tags)
        filtered_mentions[user, tag, ts] += 1
inverse_users_f = dict((v, k) for k, v in filtered_users.items())
inverse_tags_f = dict((v, k) for k, v in filtered_tags.items())
```

```python
X = sparse.DOK(shape=(len(filtered_users), len(filtered_tags), len(time)))
for user, tag, ts in filtered_mentions:
    X[filtered_users[user], filtered_tags[tag], time[ts]] = filtered_mentions[user, tag, ts]
X = X.to_coo()
```

```python
# from https://github.com/Quansight/tensorly/tree/sparse-master
import tensorly.contrib.sparse as tlsp
from tensorly.contrib.sparse.decomposition import parafac
```

```python
X = tlsp.tensor(X)
X
```

```python
A, B, C = parafac(X, 20)
```

```python
A.shape
```

```python
B.shape
```

```python
C.shape
```

```python
for factor in range(20):
    plt.plot(list(C[:, factor]))
    plt.title('Factor {}'.format(factor))
    for a in A[:, factor].todense().argsort()[::-1][:10]:
        print(inverse_users_f[a], end=' ')
    print()
    for b in B[:, factor].todense().argsort()[::-1][:10]:
        print(inverse_tags_f[b], end=' ')
    print()
    plt.show()
    print()
    print()
```
