```python
# shortest path between nodes

u=np.random.randint(0,101)
v=np.random.randint(0,101)
path = nx.shortest_path(er,source=u,target=v)

print('shortest path between {} and {} is {}'.format(u,v,path))
```

```python
pairs_dists = nx.all_pairs_shortest_path_length(er)
list(pairs_dists)
```

```python
# distances between all pairs

pairs_dists = nx.all_pairs_shortest_path_length(er)
all_dists = []
for node, dist_dict in pairs_dists:
    dist_dict.pop(node)
    all_dists = all_dists + list(dist_dict.values())

plt.hist(all_dists)
```

```python
print('Average distance: {}, theoretical: {}'.format(np.average(all_dists), np.log(101)/np.log(15)))
```

```python
# giant component

n=1000
for c in np.linspace(0.90,1.40,num=20):
    new_er = nx.erdos_renyi_graph(n, c/(n-1))
    size_gcc = max([len(cc) for cc in nx.connected_component_subgraphs(new_er)])
    print(c,size_gcc)
```

## Configuration model

We must specify a degree sequence. Here is an example using samples from a Zipf distribution.

```python
from scipy.stats import zipf

deg_seq = zipf.rvs(a=2, size=100)
deg_seq
```

```python
np.average(deg_seq)
```

```python
# sum of degrees must be odd!

cf = nx.configuration_model(deg_seq)
```

```python
# draw cf

options = {
    'node_size': 1,
    'width': 1,
}
nx.draw(cf,**options)
```
